# Análise de Dados com Python

### Desafio:

Você trabalha em uma empresa de telecom e tem clientes de vários serviços diferentes, entre os principais: internet e telefone.

O problema é que, analisando o histórico dos clientes dos últimos anos, você percebeu que a empresa está com Churn de mais de 26% dos clientes.

Isso representa uma perda de milhões para a empresa.

O que a empresa precisa fazer para resolver isso?

Base de Dados: https://drive.google.com/drive/folders/1T7D0BlWkNuy_MDpUHuBG44kT80EmRYIs?usp=sharing <br>
Link Original do Kaggle: https://www.kaggle.com/radmirzosimov/telecom-users-dataset

```python
# Passo 1: Importar a base de dados
import pandas as pd

tabela = pd.read_csv('telecom_users.csv') #variável recebe a tabela que tem como extensão o CSV (ex: ou HTML, SQL, EXCEL)

# Passo 2: Visualizar a base de dados
#display(tabela) #Exibir tabela para analisá-la
# - Entender quais as informações tão disponíveis
# - Descobrir os erros da base de dados

#Excluir colunas inúteis ao que precisamos
#axis = 0 -> eixo da linha
#axis = 1 -> eixo da coluna
tabela = tabela.drop('Unnamed: 0', axis=1) #Tabela irá receber a Tabela, dropando a coluna/linha entre os parenteses
display(tabela) #Exibir de novo para ver se deu certo.
```

```python
# Passo 3: Tratamento de dados
# print(tabela.info()) #EXIBIR: Resumo de como está sua sua Base de Dados (Valores nulos, seus tipos de dados, numero de colunas e linhas e etc...)

#Primeiro: Analisar e consertar os valores que estão reconhecidos de forma errada
tabela["TotalGasto"] = pd.to_numeric(tabela["TotalGasto"], errors="coerce") #Mudando o Tipo da coluna 'Total Gastos' para float

# Segundo: Tirando valores vazios (nulos)
# print(tabela.info()) #EXIBIR para ver onde tem valores nulos

# deletando as colunas que estão TOTALMENTE VAZIAS
# axis = 0 _> linha
# axis = 1 _> coluna
tabela = tabela.dropna(how="all", axis=1)
#colunas totoalmente vazias serão jogadas fora


# deletando as linhas que possuem em algum lugar, uma informação vazia.
#Não há problema em deletá-las, pois foi visto através do .info() que são poucas em relação ao total.
tabela = tabela.dropna(how="any", axis=0)
#se a linha tiver pelo menos UM valor vazio, será descartada

print(tabela.info())
```

```python
# Passo 4: Análise Inicial
# Como estão os nossos cancelamentos?
print(tabela["Churn"].value_counts()) #Contar quantos cancelaram ou não, utilizando a coluna 'Churn'
print(tabela["Churn"].value_counts(normalize=True).map("{:.1%}".format)) #Formato para aparecer em PORCENTAGEM
```

```python
# Passo 5: Análise Mais completa (comparar cada coluna da minha tabela com a coluna de cancelamento)
import plotly.express as px #Biblioteca para criar gráficos em Python

# etapa 1: criar o gráfico, para ficar mais fácil de visualizar as informações
for coluna in tabela.columns: #Compara as colunas com a coluna de cancelamento
    # para edições nos gráficos: https://plotly.com/python/histograms/ -> MOSTRA OS TIPOS DE GRÁFICOS
    # para mudar a cor do gráfico , color_discrete_sequence=["blue", "green"]
    grafico = px.histogram(tabela, x=coluna, color="Churn", text_auto=True)   
    # etapa 2: exibir o gráfico:
    grafico.show()
```
